Part 1: Load CSV from Virgina Department of Health.

```python
import pandas as pd
url = 'https://data.virginia.gov/resource/vsrk-d6hx.csv'
df = pd.read_csv(url)

df['rate_per_100k'] = df['infections']/df['population_denominator'] * 100000
print(df.columns.values)
```

Plot the rate of infection, based on vaccination status, utilizing the features of Week ending for temporal time frame, and the rate per 100k and vaccination status for correlation

```python
from plotnine import ggplot, aes, geom_point

(
ggplot(df, aes(x = 'week_ending_date', y = 'rate_per_100k', color = 'vaccination_status')) +
  geom_point()
)
```
